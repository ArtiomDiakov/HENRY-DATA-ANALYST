### Ahora actualizamos la columna PARTICIPANTES.

```python
import numpy as np

# Definir una función para crear el valor de 'PARTICIPANTES' basado en 'VICTIMA' y 'ACUSADO'
def update_participantes(row):
    if pd.isna(row['VICTIMA']) and pd.isna(row['ACUSADO']):
        return 'NaN-NaN'
    elif pd.isna(row['VICTIMA']):
        return f'NaN-{row["ACUSADO"]}'
    elif pd.isna(row['ACUSADO']):
        return f'{row["VICTIMA"]}-NaN'
    else:
        return f'{row["VICTIMA"]}-{row["ACUSADO"]}'

# Aplicar la función a cada fila para actualizar la columna 'PARTICIPANTES'
df['PARTICIPANTES'] = df.apply(update_participantes, axis=1)

# Verificar las primeras filas después de la actualización
print(df[['VICTIMA', 'ACUSADO', 'PARTICIPANTES']].head())
```

### Veamosla primero realizando una tabla de contingencia.

```python
import pandas as pd

# Crear la tabla de contingencia entre VICTIMA y ACUSADO
crosstab = pd.crosstab(df['VICTIMA'], df['ACUSADO'])

# Mostrar la tabla de contingencia
print(crosstab)
```

### Ahora usemos correlacion categorica (extension de chi^2) que mide la relacion de dependencia entre categorias.

```python
import numpy as np
import scipy.stats as stats

# Función para calcular Cramér's V
def cramers_v(x, y):
    confusion_matrix = pd.crosstab(x, y)
    chi2 = stats.chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    r, k = confusion_matrix.shape
    return np.sqrt(chi2 / (n * (min(r, k) - 1)))

# Calcular Cramér's V para VICTIMA y ACUSADO
cramers_v_value = cramers_v(df['VICTIMA'], df['ACUSADO'])

print(f"Cramér's V entre VICTIMA y ACUSADO: {cramers_v_value:.3f}")
```

### Podemos ver que la correlacion es cercana a 0 por lo tanto la correlacion entre victima y acusado es poca.

Vamos a analizar la distribucion de edades donde la fecha de fallecimiento exista.
Los criterios son tiene que existir EDAD en los datos y la EDAD tiene que ser mayor o igual a 1

```python
import numpy as np
import matplotlib.pyplot as plt

# Reemplazar los valores 'SD' por NaN en la columna 'EDAD'
df_victimas['EDAD'].replace('SD', np.nan, inplace=True)

# Convertir la columna 'EDAD' a números, manejando los valores no numéricos
df_victimas['EDAD'] = pd.to_numeric(df_victimas['EDAD'], errors='coerce')

# Filtrar los datos donde 'FECHA_FALLECIMIENTO' no es nula (existen datos de fallecimiento) y EDAD >= 1
victimas_fallecimiento = df_victimas[(df_victimas['FECHA_FALLECIMIENTO'].notna()) & (df_victimas['EDAD'] >= 1)]

# Verificar cuántos registros de víctimas fallecidas tenemos
print(f"Número de víctimas fallecidas con edad >= 1: {len(victimas_fallecimiento)}")

# Crear un histograma de la distribución de edades de las víctimas fallecidas
plt.figure(figsize=(10, 6))
plt.hist(victimas_fallecimiento['EDAD'].dropna(), bins=100, color='lightblue', edgecolor='black')
plt.title('Distribución de Edades de las Víctimas Fallecidas (Edad >= 1)')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.grid(True)

# Mostrar el gráfico
plt.show()

# Calcular estadísticas descriptivas de la columna 'EDAD' de las víctimas fallecidas
edad_stats = victimas_fallecimiento['EDAD'].describe()
print("\nEstadísticas descriptivas de la edad de las víctimas fallecidas:")
print(edad_stats)
```

Se debe seguramente a que los hombres son los que mas conducen los vehiculos, especialmente en edades tempranas veamos eso a continuacion.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Filtrar los datos de víctimas fallecidas con roles específicos y edades válidas
victimas_fallecimiento = df_victimas[(df_victimas['FECHA_FALLECIMIENTO'].notna()) & 
                                     (df_victimas['EDAD'] >= 1) & 
                                     (df_victimas['ROL'].isin(['CONDUCTOR', 'PASAJERO_ACOMPANANTE']))]

# Crear una tabla de contingencia para contar cuántos son CONDUCTOR o PASAJERO_ACOMPANANTE por sexo
rol_sexo_counts = pd.crosstab(victimas_fallecimiento['ROL'], victimas_fallecimiento['SEXO'])

# Mostrar la tabla de contingencia
print("Conteo de roles por sexo:")
print(rol_sexo_counts)

# Crear un gráfico de barras para visualizar la relación entre ROL y SEXO
rol_sexo_counts.plot(kind='bar', figsize=(10, 6), color=['lightblue', 'lightgreen'])
plt.title('Relación entre ROL (CONDUCTOR vs PASAJERO_ACOMPANANTE) y SEXO')
plt.xlabel('ROL')
plt.ylabel('Cantidad')
plt.xticks(rotation=0)
plt.legend(title='Sexo')
plt.grid(True)

# Mostrar el gráfico
plt.show()
```

Parece que la hipotesis es cierta se puede notar que victimas hombres conductores son mucho mayores que mujeres.

Realizemos ahora la distribucion de edades de conductores victimas por sexo.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Filtrar los datos para obtener únicamente los conductores (ROL = 'CONDUCTOR')
conductores = victimas_fallecimiento[victimas_fallecimiento['ROL'] == 'CONDUCTOR']

# Crear un gráfico de violín para visualizar la distribución de edades de los conductores por sexo
plt.figure(figsize=(10, 6))
sns.violinplot(x='SEXO', y='EDAD', data=conductores, palette='Set2')
plt.title('Distribución de Edades de Conductores Víctimas por Sexo')
plt.xlabel('Sexo')
plt.ylabel('Edad')
plt.grid(True)

# Mostrar el gráfico
plt.show()

# También podemos agregar un boxplot para una vista alternativa de la distribución
plt.figure(figsize=(10, 6))
sns.boxplot(x='SEXO', y='EDAD', data=conductores, palette='Set3')
plt.title('Boxplot de Edades de Conductores Víctimas por Sexo')
plt.xlabel('Sexo')
plt.ylabel('Edad')
plt.grid(True)

# Mostrar el gráfico
plt.show()
```

Como podemos observar ahora la distribucion es casi igual con media de casi de 32 anios para los que son conductores en ambos sexos.

Aun asi es preocupante en hombres cercanos a 25 anios de edad. Seguramente por que son jovenes con muchas ganas de vivir y tambien por abuso excesivo de alcohol pero estos datos no tenemos para determinar a ciencia cierta.
